## 1. Getting Data
We will first get our data and normalize. For this **Alpha** version we only have ```MNIST``` dataset. For future development we will include more

````python
# Grabbing raw data: mapped to tensors
x_train, y_train, x_valid, y_valid = get_data()
````

````python
# Normalizing our data
??normalize_to
````

````python
x_train, x_valid = normalize_to(x_train, x_valid)
````

````python
# checking stats
x_train.mean(), x_train.std()
````

````python
# creating datasets
??Dataset
````

````python
train_ds, valid_ds = Dataset(x_train, y_train), Dataset(x_valid, y_valid)
````

````python
# grabbing dataloaders
??get_dls
````

````python
bs = 512
train_dl, valid_dl = get_dls(train_ds, valid_ds, bs)
````

````python
# Creating DataBunch object
??DataBunch
````

````python
c = y_train.max().item()+1
data = DataBunch(train_dl, valid_dl, c)
````

## 2. Callbacks
These are callbacks that are called at various points of the model. Check out the course or my rep for more on this

````python
mnist_view = view_tfms(1,28,28) # will transform our model at beginning of training
cbfs = [Recorder, CudaCallback]
cbfs.append(partial(AvgStatsCallback, accuracy))
cbfs.append(partial(BatchTransformXCallback, mnist_view))
````

````python
# example od our transform
original_shape = x_train[1].shape
batchtransform = Lambda(view_tfms(1,28,28))
new_shape = batchtransform(x_train[1])

original_shape, new_shape.shape
````
